```python
df_cohort = copy.deepcopy(df_transactions_dc_merge)
```

```python
df_cohort['Invoice Date'] = pd.to_datetime(df_cohort['Invoice Date'], utc=True).dt.tz_localize(None)
df_cohort['CohortMonth'] = df_cohort.groupby('Customer ID')['Invoice Date'].transform('min').dt.to_period('M')
df_cohort['PurchaseMonth'] = df_cohort['Invoice Date'].dt.to_period('M')
df_cohort['CohortIndex'] = (df_cohort['PurchaseMonth'] - df_cohort['CohortMonth']).apply(lambda x: x.n)
```

```python
df_cohort.head()
df_cohort.to_excel('df_cohort.xlsx')

```

```python
# Group by CohortMonth and CohortIndex, then count the number of unique customers
cohort_data = df_cohort.groupby(['CohortMonth', 'CohortIndex']).agg({
    'Customer ID': 'nunique'
}).reset_index()

# Pivot the data so that each row represents a cohort and each column represents the months since first purchase
cohort_pivot = cohort_data.pivot_table(
    index='CohortMonth',
    columns='CohortIndex',
    values='Customer ID'
)

# Calculate retention by dividing each column by the first column (initial cohort size)
cohort_size = cohort_pivot.iloc[:, 0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)

# Display the retention matrix
retention_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.heatmap(retention_matrix, annot=True, fmt='.0%', cmap='Blues')
plt.title('Cohort Retention Rate')
plt.show()
```

```python
# Calculate the total revenue by cohort
cohort_revenue = df_cohort.groupby(['CohortMonth', 'CohortIndex']).agg({
    'Amount': 'sum'
}).reset_index()

# Pivot the revenue data for better analysis
revenue_pivot = cohort_revenue.pivot_table(
    index='CohortMonth',
    columns='CohortIndex',
    values='Amount'
)

# Visualize the revenue matrix
plt.figure(figsize=(12, 8))
sns.heatmap(revenue_pivot, annot=True, fmt='.1f', cmap='Greens')
plt.title('Cohort Revenue')
plt.show()
```
